###Adding Additional Features

###SMA - Simple Moving Average
It is the average of the price of the stock over a period of time

Its formula Being:
                   sum(price for n days)/n days

```python
def SMA(df,n):
  for i in range(len(df)):
    sma=sum(df.Price[i:i+n])/n
    df.loc[i,'SMA']=sma
```

###RSI - Relative Strenght Index

The RSI is typically calculated using a 14-day period. The formula is:

RSI = 100 - [100 / (1 + RS)]

Where RS = Average Gain / Average Loss


```python
def avg_gain_loss(df):
  for i in range(len(df)):
    while (i<len(df)-14):
      gain=0
      loss=0
      for j in range(i,i+14):
        if df.Change[i]>0:
          gain+=df.Change[i]
        else:
          loss+=df.Change[i]
      avg_gain=gain/14
      avg_loss=loss/14
      df.loc[i,'avg_gain']=avg_gain
      df.loc[i,'avg_loss']=avg_loss
      i+=1
```

```python
def avg_gain_loss1(df):
  for i in range(len(df)):
    while (i<len(df)-14):
      gain=0
      loss=0
      gain=sum(df.Change[i:i+14][df.Change[i:i+14]>0])
      loss=sum(df.Change[i:i+14][df.Change[i:i+14]<0])
      avg_gain=gain/14
      avg_loss=loss/14
      df.loc[i,'avg_gain']=avg_gain
      df.loc[i,'avg_loss']=avg_loss
      i+=1
```

```python
def avg_gain_loss2(df):
    # Initialize columns for average gain and average loss
    df['avg_gain'] = 0.0
    df['avg_loss'] = 0.0

    # Calculate initial sums for first 14 days
    initial_window = df.loc[0:13, 'Change']
    initial_gain = initial_window[initial_window > 0].sum()
    initial_loss = abs(initial_window[initial_window < 0].sum())

    # Set initial average gains and losses
    df.loc[13, 'avg_gain'] = initial_gain / 14
    df.loc[13, 'avg_loss'] = initial_loss / 14

    # Calculate average gain and loss for subsequent days
    for i in range(14, len(df)):
        current_gain = df.loc[i, 'Change'] if df.loc[i, 'Change'] > 0 else 0
        current_loss = abs(df.loc[i, 'Change']) if df.loc[i, 'Change'] < 0 else 0

        # Update average gain and loss using previous day's values
        df.loc[i, 'avg_gain'] = (df.loc[i-1, 'avg_gain'] * 13 + current_gain) / 14
        df.loc[i, 'avg_loss'] = (df.loc[i-1, 'avg_loss'] * 13 + current_loss) / 14


    return df
```
